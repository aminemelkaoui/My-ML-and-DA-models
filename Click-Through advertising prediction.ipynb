```python
import pandas as pd
data = pd.read_csv('advertising.csv')
data.head()
```

```python
data.info()
```

```python
data.columns
```

```python
train=data.iloc[:,0:7]
train=train.drop(['Ad Topic Line','City'],axis=1)
train
```

```python
target=data.iloc[:,9]
target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(train,target,test_size=0.4,random_state=4)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
Lr=LogisticRegression(C=0.03,random_state=0)
Lr.fit(x_train,y_train)
y_pred=Lr.predict(x_test)
print(y_pred)
```

```python
y_pred_proba=Lr.predict_proba(x_test)
y_pred_proba
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import f1_score
print(f1_score(y_test,y_pred))
```
